#### Add Packages to Access Worksheet Data

```python
# import packages to work with Google Sheets
from google.colab import auth
import gspread
from google.auth import default

#authenticating to google
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

#### Read Worksheet Data in as pd.DataFrame and Adjust Indices

```python
# open the worksheet and save as a dataframe
worksheet = gc.open('Data Integration and Reporting Sandbox (Lever API)').worksheet('Postings_Report')

# get_all_values gives a list of rows
rows = worksheet.get_all_values()

# convert to a DataFrame
df = pd.DataFrame(rows[1:], columns=rows[0])
df.set_index('Posting ID', inplace=True)
```
